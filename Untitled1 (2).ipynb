```python
# Import required libraries (if not already imported)
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Compute WCSS for different numbers of clusters
wcss = []
max_clusters = 10  # Define max clusters for reusability

for i in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)  # Explicitly set n_init to avoid warning
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method
plt.figure(figsize=(8, 5))  # Set figure size for better readability
plt.plot(range(1, max_clusters + 1), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal Number of Clusters', pad=20)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid for better readability
plt.xticks(range(1, max_clusters + 1))  # Ensure x-axis ticks are integers
plt.show()
```
